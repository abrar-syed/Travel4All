```
req = Request("https://gasprices.aaa.com/state-gas-price-averages/", headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = bs(webpage, features="html.parser")
table = soup.find(id='sortable')
```

```
# extract all the states included in table
states = []
for s in table.find_all('a'):
  state = s.text[2:].strip()
  states.append(state)
print(states)
```

```
# extract all the regular gas prices included in table
regular_prices = []

# exclude the first element in find_all list (header)
for reg in table.find_all(class_='regular')[1:]:
  price = reg.text
  # get rid of dollar sign and surrounding white spaces
  price = reg.text.replace("$", "").strip()

  #convert to number
  price = float(price)
  regular_prices.append(price)
print(regular_prices)
```

```
# extract all the mid_grade gas prices included in table
mid_prices = []

# exclude the first element in find_all list (header)
for reg in table.find_all(class_='mid_grade')[1:]:
  price = reg.text
  # get rid of dollar sign and surrounding white spaces
  price = reg.text.replace("$", "").strip()

  #convert to number
  price = float(price)
  mid_prices.append(price)
print(mid_prices)
```

```
# extract all the premium gas prices included in table
prem_prices = []

# exclude the first element in find_all list (header)
for reg in table.find_all(class_='premium')[1:]:
  price = reg.text
  # get rid of dollar sign and surrounding white spaces
  price = reg.text.replace("$", "").strip()

  #convert to number
  price = float(price)
  prem_prices.append(price)
print(prem_prices)
```

```
# extract all the diesel gas prices included in table
d_prices = []

# exclude the first element in find_all list (header)
for reg in table.find_all(class_='diesel')[1:]:
  price = reg.text
  # get rid of dollar sign and surrounding white spaces
  price = reg.text.replace("$", "").strip()

  #convert to number
  price = float(price)
  d_prices.append(price)
print(d_prices)
```

```
d = {'State':states, 'Regular Price':regular_prices, 'Mid-Grade Price':mid_prices, 'Premium Price':prem_prices, 'Diesel Price':d_prices}
gas_prices = pd.DataFrame(data = d)
gas_prices
```

```
#Gas Price Script
#function calculates average gas price for a trip that starts in a given state based on fuel type, trip length, and
#city/highway mpg
#city_mpg and highway_mpg are user inputs

def calc_gas_price(state, city_mpg, highway_mpg, fuel_type, trip_length):
    cost_per_gallon = gas_prices[gas_prices['State'] == state][fuel_type]
    avg_mpg = 0.5*(city_mpg + highway_mpg)
    price = (trip_length/avg_mpg)*cost_per_gallon
    return price  
```
